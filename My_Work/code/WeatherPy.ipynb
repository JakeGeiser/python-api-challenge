```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from pprint import pprint

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []
countries = []

test_size = 1500
# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=test_size)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=test_size)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    near_city = citipy.nearest_city(lat_lng[0], lat_lng[1])
    city = near_city.city_name
    country = near_city.country_code
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        countries.append(country)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# define the base url link for the current weather API
base_url = "https://api.openweathermap.org/data/2.5/weather?"
# Create an empty dictionary for storing successful requests responses in
weather_d = {
    'City':[],
    'Lat':[],
    'Lng':[],
    'Max Temp(F)':[],
    'Humidity':[],
    'Cloudiness':[],
    'Wind Speed':[],
    'Country':[],
    'Date':[]
}
removed_cities = []
removed_countries = []

for i in range(0,len(cities)):
    try:
        params = f'q={cities[i]},{countries[i]}&units=imperial&appid={weather_api_key}'
        response = requests.get(base_url + params).json()
        weather_d['Lat'].append(response['coord']['lat'])
        weather_d['Lng'].append(response['coord']['lon'])
        weather_d['Max Temp'].append(response['main']['temp_max'])
        weather_d['Humidity'].append(response['main']['humidity'])
        weather_d['Cloudiness'].append(response['clouds']['all'])
        weather_d['Wind Speed'].append(response['wind']['speed'])
        weather_d['Country'].append(countries[i].upper())
        weather_d['Date'].append(response['dt'])
        weather_d['City'].append(cities[i].capitalize())
        print(f'Processing Record index {i} | {cities[i]}, {countries[i]}')
    except:
        removed_cities.append(cities[i])
        removed_countries.append(cities[i])
        print(f'City( {cities[i]}, {countries[i]}) not found. Skipping...')
        
```

### Convert Raw Data to DataFrame

```python
# get a weather dataframe and store as a csv file
weather_df = pd.DataFrame(weather_d)
weather_df.to_csv("../output_data/city_weather")
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
def lin_plot(df,y_axis,hemisphere):
    # get linear regression stats
    slope, intercept, r_value, p_value, std_err = linregress(df['Lat'], df[y_axis])
    # plot data
    plt.figure()
    df.plot(kind='scatter',x='Lat',y=f'{y_axis}',alpha=0.60,edgecolor='black')
    plt.xlabel('Latitude')
    plt.ylabel(f'{y_axis}')
    plt.title(f'{hemisphere} Hemisphere City {y_axis} vs Latitude (07/27/2020)')
    # plot linear regression
    if hemisphere == 'Northern':
        lats = np.arange(0,max(df['Lat']))
    if hemisphere == 'Southern':
        lats = np.arange(min(df['Lat']),0)
    plt.plot(lats,lats*slope + intercept,'r-',label=f'{slope:.2f}x + {intercept:.2f}')
    plt.legend()
    # print out statisitcs
    print(f"Correlation coefficent: {r_value}")
    print(f"P value: {p_value}")
    # save the figure
    plt.savefig(f"../output_data/{hemisphere}_{y_axis.replace(' ','')}.png")
```

```python
# Create Northern and Southern Hemisphere DataFrames
north_df = minimal_data_df.loc[minimal_data_df['Lat']>0,:]
south_df = minimal_data_df.loc[minimal_data_df['Lat']<0,:]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
lin_plot(north_df,'Max Temp','Northern')
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
lin_plot(south_df,'Max Temp','Southern')
```

The southern and northern max temp vs latitude show significant correlation. The northern has a R value of -0.69 and the southern has a R value of 0.80. This indicates that the southern hemisphere data used fits the linear slope better than the northern does. I would suggests that the temperature is acutally the 'hottest' around 25 degrees latitude. This is due to the Earth's tilt and the seasons it experiences. During the northern summer, the northern hemisphere is facing the sun more, thus higher temperature. A better analysis would to use a temperature average for each city to avoid the seasons being a factor. But the historical weather api cost money when I looked into it, so I used the daily.
